```python
import numpy as np
import matplotlib.pyplot as plt

import scipy.sparse as sp
import pickle, time, os
```

```python
"""
a1 = np.array([0, 2])
a2 = np.array([2, 0])


unit_cell_vecs = np.array([[0, 0],
                           [0, 1],
                           [1, 0],
                           [1, 1]])
"""
a1 = np.array([1, 1])
a2 = np.array([1, -1])


unit_cell_vecs = np.array([[0, 0],
                           [0, 1]])
```

```python
def generate_supercell_vectors(Lx, Ly):
    R_vec = np.zeros(shape=(unit_cell_vecs.shape[0]*Lx*Ly, 2))
    for i in range(Lx):
        for j in range(Ly):
            for k, d in enumerate(unit_cell_vecs):
                R_vec[unit_cell_vecs.shape[0]*(i+Lx*j)+k, :] = i*a1+j*a2+d
            
    return R_vec
```

```python
Lx, Ly = 2, 2

R_list = generate_supercell_vectors(Lx,Ly)
for i in range(unit_cell_vecs.shape[0]*Ly*Lx):
    plt.scatter(R_list[i, 0], R_list[i, 1], c='k')
```
